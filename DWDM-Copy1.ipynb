```python
import pandas as pd
import numpy as np
data=pd.read_csv("Salary_Data.csv")
display(data)
```

```python
x=data.iloc[:,:-1]
y=data.iloc[:,-1]
display(x)
display(y)
```

```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
lg=LinearRegression()
lg.fit(x_train,y_train)
y_pred=lg.predict(x_test)
plt.scatter(x_train,y_train,color='orange')
plt.plot(x_train,lg.predict(x_train),color='blue')
plt.title("Salary vs Experience")
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.show()
plt.scatter(x_test,y_test,color='red')
plt.plot(x_test,lg.predict(x_test),color='blue')
plt.title("Salary vs Experience")
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.show()
```

```python
import pandas as pd
df=pd.read_csv(r"Mall_cust.csv")
display(df)
```

```python
x=df.iloc[:,[3,4]].values
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
kmeans=KMeans(n_clusters=5)
y_predict=kmeans.fit_predict(x)
plt.scatter(x[y_predict==0,0],x[y_predict==0,1],s=100,color='red',label='cluster1')
plt.scatter(x[y_predict==1,0],x[y_predict==1,1],s=100,color='cyan',label='cluster2')
plt.scatter(x[y_predict==2,0],x[y_predict==2,1],s=100,color='orange',label='cluster3')
plt.scatter(x[y_predict==3,0],x[y_predict==3,1],s=100,color='purple',label='cluster4')
plt.scatter(x[y_predict==4,0],x[y_predict==4,1],s=100,color='green',label='cluster5')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=300,color='yellow',label='centroid')
plt.title("Clusters of Customers")
plt.xlabel("Annual Income")
plt.ylabel("Spending Score(1-100)")
plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
df = pd.read_csv(r"Mall_Cust.csv")
X_train = df[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']]
clustering = DBSCAN(eps=12.5, min_samples=4).fit(X_train)
DBSCAN_dataset = X_train.copy()
DBSCAN_dataset.loc[:,'Cluster'] = clustering.labels_
DBSCAN_dataset.Cluster.value_counts().to_frame()
outliers = DBSCAN_dataset[DBSCAN_dataset['Cluster']==-1]
fig2, (axes) = plt.subplots(1,2,figsize=(12,5))
sns.scatterplot('Annual Income(k$)','Spending Score (1-100)',data=DBSCAN_dataset[DBSCAN_dataset['Cluster']!=-1],hue='Cluster', ax=axes[0], palette='Set2',legend='full',s=200)
sns.scatterplot('Age', 'Spending Score (1-100)',data=DBSCAN_dataset[DBSCAN_dataset['Cluster']!=-1],hue='Cluster', palette='Set2', ax=axes[1], legend='full',s=200)
axes[0].scatter(outliers['Annual Income (k$)'], outliers['Spending Score (1-100)'], s=10, label='outliers', c="k")
axes[1].scatter(outliers['Age'], outliers['Spending Score (1-100)'], s=10,label='outliers', c="k")
axes[0].legend()
axes[1].legend()
plt.setp(axes[0].get_legend().get_texts(), fontsize='12')
plt.setp(axes[1].get_legend().get_texts(), fontsize='12')
plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN

# Load dataset
df = pd.read_csv(r"Mall_Cust.csv")

# Select features for clustering
X_train = df[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']]

# Apply DBSCAN clustering
clustering = DBSCAN(eps=12.5, min_samples=4).fit(X_train)

# Create a copy of the dataset with cluster labels
DBSCAN_dataset = X_train.copy()
DBSCAN_dataset['Cluster'] = clustering.labels_

# Identify outliers
outliers = DBSCAN_dataset[DBSCAN_dataset['Cluster'] == -1]

# Set up the plotting area
fig2, axes = plt.subplots(1, 2, figsize=(12, 5))

# Scatter plot for 'Annual Income' vs 'Spending Score'
sns.scatterplot(
    x='Annual Income (k$)', y='Spending Score (1-100)', 
    data=DBSCAN_dataset[DBSCAN_dataset['Cluster'] != -1],
    hue='Cluster', ax=axes[0], palette='Set2', legend='full', s=200
)
axes[0].scatter(
    outliers['Annual Income (k$)'], outliers['Spending Score (1-100)'], 
    s=10, label='outliers', c="k"
)

# Scatter plot for 'Age' vs 'Spending Score'
sns.scatterplot(
    x='Age', y='Spending Score (1-100)', 
    data=DBSCAN_dataset[DBSCAN_dataset['Cluster'] != -1],
    hue='Cluster', palette='Set2', ax=axes[1], legend='full', s=200
)
axes[1].scatter(
    outliers['Age'], outliers['Spending Score (1-100)'], 
    s=10, label='outliers', c="k"
)

# Add legends
axes[0].legend()
axes[1].legend()

# Adjust legend text size
plt.setp(axes[0].get_legend().get_texts(), fontsize='12')
plt.setp(axes[1].get_legend().get_texts(), fontsize='12')

# Display the plots
plt.show()
```
